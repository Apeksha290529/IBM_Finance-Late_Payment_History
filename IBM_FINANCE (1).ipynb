```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cols_date = ["PaperlessDate","InvoiceDate","DueDate","SettledDate"]
for col in cols_date:
    df[col] = pd.to_datetime(df[col],format='%m/%d/%Y')
df.head()
```

```python
df.customerID.value_counts()
```

```python
temp = df.groupby(["countryCode"], axis=0, as_index=False)["customerID"].count()
sns.barplot(x='countryCode',y='customerID',data=temp,palette='bright')
plt.title('Customers per country')
```

```python
temp = df.groupby(["customerID"], axis=0, as_index=False)["InvoiceAmount"].mean().sort_values(by=['InvoiceAmount'],ascending=False)
temp.head(5)
```

```python
plt.figure(figsize=(25,5))
sns.barplot(x='customerID',y='InvoiceAmount',data=temp,palette='bright')
plt.title('Average InvoiceAmount per Customer')
_ = plt.xticks(rotation=90,fontsize=14)
```

```python
temp = df.groupby(["customerID"], axis=0, as_index=False)["DaysToSettle"].mean().sort_values(by=['DaysToSettle'],ascending=False)
temp.head(5)
```

```python
plt.figure(figsize=(25,5))
sns.barplot(x='customerID',y='DaysToSettle',data=temp,palette='bright')
plt.title('Average DaysToSettle per Customer')
_ = plt.xticks(rotation=90,fontsize=14)
```

```python
temp = df.groupby(["customerID"], axis=0, as_index=False)["DaysLate"].mean().sort_values(by=['DaysLate'],ascending=False)
temp.head(5)
```

```python
temp.tail(20)
```

```python
plt.figure(figsize=(25,5))
sns.barplot(x='customerID',y='DaysLate',data=temp,palette='bright')
plt.title('Average DaysLate per Customer')
_ = plt.xticks(rotation=90,fontsize=14)
```

```python
temp = df.groupby(["countryCode"], axis=0, as_index=False)[["InvoiceAmount","DaysToSettle","DaysLate"]].mean()
print('Average InvoiceAmount, DaysToSettle and DaysLate per country:')
temp.head(5)
```

```python
temp = pd.DataFrame(df.groupby(['countryCode'], axis=0, as_index=False)['DaysLate'].mean())
plt.figure(figsize=(10,6))
sns.barplot(x="countryCode", y="DaysLate",data=temp,linewidth=2.5, facecolor=(1, 1, 1, 0),
                 errcolor=".4", edgecolor="red")
```

```python
temp = df.groupby(["InvoiceMonth"], axis=0, as_index=False)["invoiceNumber"].count()
sns.barplot(x='InvoiceMonth',y='invoiceNumber',data=temp,palette='bright')
plt.title('Invoices per month')
_ = plt.xticks(rotation=90,fontsize=14)
```
